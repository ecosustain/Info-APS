```python
import pandas as pd
import chardet

```

```python
# Ler alguns bytes do arquivo para detectar a codificação
with open('data/populacao.csv', 'rb') as f:
    result = chardet.detect(f.read(100000))
result
```

```python
df = pd.read_csv('data/populacao.csv', encoding='utf-8', sep=';')
df
```

```python
df.iloc[16]
```

```python
# Renomear colunas
df.columns = ['uf', 'uf_cod', 'ibge_cod', 'municipio', 'populacao']
```

```python
# Separar coluna populacao pegando apenas o primeiro valor antes do parenteses
df['populacao'] = df['populacao'].str.split('(').str[0]

# remover pontos
df['populacao'] = df['populacao'].str.replace('.', '')

# converter para inteiro
df['populacao'] = df['populacao'].astype(int)
df
```

```python
df.to_csv('data/populacao.csv', index=False)
```
