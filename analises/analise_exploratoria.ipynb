```python
# Plotar o histograma da variavel Total em escala logaritmica
df['Total'].plot(kind='hist', bins=50, log=True)
```

```python
# Plotar o histograma da variavel Total
df['Total'].plot(kind='hist')
```

```python
# Calcular os quartis
q1 = df['Total'].quantile(0.25)
median = df['Total'].quantile(0.5)
q3 = df['Total'].quantile(0.75)

# Criar o boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(y='Total', data=df)
plt.yscale('log')

# Adicionar os valores dos quartis ao gráfico
plt.text(0.6, q1, f'Q1: {q1:.2f}', ha='left', va='center', fontsize=12)
plt.text(0.6, median, f'Mediana: {median:.2f}', ha='left', va='center', fontsize=12)
plt.text(0.6, q3, f'Q3: {q3:.2f}', ha='left', va='center', fontsize=12)

plt.show()
```

```python
# Agrupar os dados por Região e somar os valores de Total
df.groupby('Região').agg({'Total': ['mean', 'sum', 'count']})
```

```python
# Estados e Competencia com maior valor de Total
df.groupby(['Uf', 'Mes', 'Ano']).sum().nlargest(10, 'Total')
```

```python
# Região e Competencia com maior valor de Total
df.groupby(['Região', 'Mes', 'Ano']).sum().nlargest(10, 'Total')
```

```python
# Municipio e Competencia com maior valor de Total
df.groupby(['Municipio', 'Mes', 'Ano']).sum().nlargest(5, 'Total')
```

```python
# Ano e Mes com maior valor de Total
df.groupby(['Ano', 'Mes']).sum().nlargest(5, 'Total')
```
